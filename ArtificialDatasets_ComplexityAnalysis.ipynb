## Complexity boxplots

```python
path_csv = os.chdir(root_path+'/Results_Complexity_InstanceLevel')


# Extraemos los nombres de todos los ficheros
total_name_list = []
for filename in os.listdir(path_csv):
    if filename.endswith('.csv'):
        total_name_list.append(filename)

total_name_list = ['ComplexityMeasures_InstanceLevel_Dataset2.csv','ComplexityMeasures_InstanceLevel_Dataset3.csv',
                   'ComplexityMeasures_InstanceLevel_Dataset6.csv','ComplexityMeasures_InstanceLevel_Dataset7.csv',
                   'ComplexityMeasures_InstanceLevel_Dataset8.csv','ComplexityMeasures_InstanceLevel_Dataset10.csv',
                   'ComplexityMeasures_InstanceLevel_Dataset11.csv','ComplexityMeasures_InstanceLevel_Dataset14.csv',
                   'ComplexityMeasures_InstanceLevel_Moon2.csv',
                   'ComplexityMeasures_InstanceLevel_Dataset17.csv','ComplexityMeasures_InstanceLevel_Dataset18.csv',
                   'ComplexityMeasures_InstanceLevel_Dataset19.csv','ComplexityMeasures_InstanceLevel_Dataset20.csv',
                   'ComplexityMeasures_InstanceLevel_Dataset21.csv']

cols_name = ['dataset','Hostility', 'kDN', 'DS', 'DCP', 'TD_U', 'TD_P', 'MV', 'CB', 'CLD',
       'N1', 'N2', 'LSC', 'LSradius', 'H', 'U', 'F1', 'F2', 'F3', 'F4']
data_total = pd.DataFrame(columns=cols_name)

for data_file in total_name_list:
    # print(data_file)
    file = data_file

    file_name = data_file[33:-4]
    data = pd.read_csv(file,index_col=0)
    data['dataset'] = file_name
    data_total = pd.concat([data_total, data], ignore_index=True)

```

```python
# order_fixed=['Dataset1','Dataset2','Dataset3','Dataset4','Dataset5','Dataset6','Dataset7','Dataset8','Dataset9',
#        'Dataset10','Dataset11','Dataset12','Dataset13','Dataset14','Dataset15','Dataset16','Moon','Moon2','XOR']
order_fixed=['Dataset2','Dataset3','Dataset6','Dataset7','Dataset8',
       'Dataset10','Dataset11','Dataset14','Dataset17','Dataset18','Dataset19','Dataset20',
             'Dataset21','Moon2']
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 18))

# fig.suptitle('Suptitle')
sns.boxplot(ax=axes[0, 0],data=data_total, x="CLD", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[0, 1],data=data_total, x="DCP", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[0, 2],data=data_total, x="F1", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[1, 0],data=data_total, x="LSC", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[1, 1],data=data_total, x="N1", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[1, 2],data=data_total, x="N2", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[2, 0],data=data_total, x="TD_U", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[2, 1],data=data_total, x="kDN", y="dataset", order=order_fixed)
sns.boxplot(ax=axes[2, 2],data=data_total, x="Hostility", y="dataset", order=order_fixed)
plt.show()
```
